```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
```

```python
print("Chargement du modèle FinBERT...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")

# 3. Créer une fonction pour analyser le sentiment
def analyse_sentiment_financier(texte):
    """
    Analyse le sentiment d'un texte financier en utilisant FinBERT.

    Args:
        texte (str): Le texte financier à analyser.

    Returns:
        dict: Un dictionnaire contenant les résultats du sentiment et les probabilités.
    """
    # Pipeline pour l'analyse des sentiments
    nlp_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
    resultat = nlp_pipeline(texte)
    return resultat
```

```python
# 4. Exemple d'utilisation
if __name__ == "__main__":
    # Exemple de texte financier
    texte_financier = """
    The company's stock price increased significantly after reporting higher-than-expected earnings for Q4.
    However, concerns remain about inflation and its impact on future growth.
    """

    print("Analyse du sentiment pour le texte suivant :")
    print(texte_financier)

    # Analyse du texte
    resultat = analyse_sentiment_financier(texte_financier)

    # Afficher les résultats
    print("\nRésultats de l'analyse des sentiments :")
    for sentiment in resultat:
        print(f"- Label : {sentiment['label']}, Score : {sentiment['score']:.2f}")
```

```python
# 4. Interface pour demander un texte à l'utilisateur
if __name__ == "__main__":
    print("Bienvenue dans l'analyseur de sentiments financiers avec FinBERT !")
    print("Veuillez entrer un texte financier à analyser (par exemple, un rapport ou une phrase) :")

    # L'utilisateur entre son texte
    texte_financier = input("\nVotre texte : ")

    # Vérifier si l'utilisateur a fourni un texte valide
    if texte_financier.strip() == "":
        print("Erreur : Aucun texte fourni. Veuillez réessayer avec un texte valide.")
    else:
        print("\nAnalyse du sentiment en cours...")

        # Analyse du texte
        resultat = analyse_sentiment_financier(texte_financier)

        # Afficher les résultats
        print("\nRésultats de l'analyse des sentiments :")
        for sentiment in resultat:
            print(f"- Label : {sentiment['label']}, Score : {sentiment['score']:.2f}")
```

# ***Classification des textes financiers en catégories***

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# 2. Charger le modèle FinBERT pour une tâche de classification personnalisée
print("Chargement du modèle FinBERT...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")

# 3. Définir les catégories personnalisées (labels)
categories = ["risque", "opportunité", "neutre"]
```

```python
# 4. Fonction pour classifier un texte financier
def classifier_texte_financier(texte):
    """
    Classifie un texte financier en catégories : risque, opportunité, ou neutre.

    Args:
        texte (str): Le texte financier à analyser.

    Returns:
        dict: Un dictionnaire contenant les résultats de classification.
    """
    # Pipeline pour la classification
    nlp_pipeline = pipeline("text-classification", model=model, tokenizer=tokenizer, return_all_scores=True)

    # Analyse du texte
    resultat = nlp_pipeline(texte)

    # Associer les catégories personnalisées aux scores
    resultat_categorise = {categories[i]: score['score'] for i, score in enumerate(resultat[0])}

    return resultat_categorise
```

```python
# 5. Exemple d'utilisation
if __name__ == "__main__":
    print("Bienvenue dans le classificateur de textes financiers avec FinBERT !")
    print("Veuillez entrer un texte financier à analyser :")

    # L'utilisateur entre un texte financier
    texte_financier = input("\nVotre texte : ")

    # Vérifier si un texte valide est fourni
    if texte_financier.strip() == "":
        print("Erreur : Aucun texte fourni. Veuillez réessayer avec un texte valide.")
    else:
        print("\nClassification du texte en cours...")

        # Classifier le texte
        resultat = classifier_texte_financier(texte_financier)

        # Afficher les résultats
        print("\nRésultats de la classification :")
        for categorie, score in resultat.items():
            print(f"- Catégorie : {categorie}, Score : {score:.2f}")
```

# **Extraction d'entités nommées (NER) dans les textes financiers**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Étape 2 : Charger FinBERT
# FinBERT est adapté aux textes financiers.
print("Chargement de FinBERT pour l'extraction d'entités sentimentales...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")

# Étape 3 : Configurer un pipeline pour analyser les sentiments d'entités
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
# Étape 4 : Fonction pour classer le sentiment des phrases contenant des entités
def analyser_sentiment_entites(texte):
    """
    Analyse le sentiment des entités financières dans un texte.

    Args:
        texte (str): Le texte financier à analyser.

    Returns:
        list: Une liste des entités avec leur sentiment.
    """
    phrases = texte.split(". ")  # Diviser le texte en phrases
    resultats = []
    for phrase in phrases:
        analyse = sentiment_pipeline(phrase)
        resultats.append({"phrase": phrase, "sentiment": analyse[0]['label'], "score": analyse[0]['score']})
    return resultats
```

```python
# Étape 5 : Utilisation du modèle
if __name__ == "__main__":
    print("Extraction d'entités et analyse de sentiment avec FinBERT.")

    # Demander un texte financier à l'utilisateur
    texte_financier = input("Entrez un texte financier : ")

    if texte_financier.strip() == "":
        print("Erreur : Aucun texte fourni.")
    else:
        print("\nAnalyse des entités en cours...")
        resultat = analyser_sentiment_entites(texte_financier)

        # Afficher les résultats
        print("\nRésultats d'analyse des entités :")
        for r in resultat:
            print(f"Phrase : {r['phrase']}\nSentiment : {r['sentiment']} (Score : {r['score']:.2f})\n")
```

# **Résumé des textes financiers**

```python
# Étape 1 : Importer les bibliothèques nécessaires
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Étape 2 : Charger FinBERT
print("Chargement de FinBERT pour résumé contextuel...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")

# Étape 3 : Configurer un pipeline pour analyser les sentiments
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Étape 4 : Fonction pour générer un résumé basé sur les sentiments
def resumer_texte_financier(texte):
    """
    Résume un texte financier en priorisant les phrases clés.

    Args:
        texte (str): Le texte financier à résumer.

    Returns:
        str: Un résumé des phrases importantes.
    """
    phrases = texte.split(". ")  # Diviser le texte en phrases
    phrases_importantes = []

    for phrase in phrases:
        analyse = sentiment_pipeline(phrase)[0]
        if analyse['label'] in ['POSITIVE', 'NEGATIVE']:  # On retient les phrases au ton marqué
            phrases_importantes.append(phrase)

    return " ".join(phrases_importantes)

# Étape 5 : Utilisation du modèle
if __name__ == "__main__":
    print("Résumé contextuel avec FinBERT.")

    # Demander un texte financier à l'utilisateur
    texte_financier = input("Entrez un texte financier : ")

    if texte_financier.strip() == "":
        print("Erreur : Aucun texte fourni.")
    else:
        print("\nGénération du résumé en cours...")
        resultat = resumer_texte_financier(texte_financier)

        # Afficher le résumé
        print("\nRésumé généré :")
        print(resultat)
```

# **Réponses aux questions spécifiques à un texte financier avec FinBERT**

```python
# 1. Importer les bibliothèques nécessaires
# !pip install transformers torch

from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline

# 2. Charger un modèle de Question Answering (FinBERT ou similaire)
print("Chargement du modèle pour les réponses aux questions...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForQuestionAnswering.from_pretrained("yiyanghkust/finbert-tone")

# 3. Créer un pipeline pour le Question Answering
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)

# 4. Fonction pour répondre à une question spécifique
def repondre_a_question(texte, question):
    """
    Répond à une question basée sur un texte donné.

    Args:
        texte (str): Le texte source.
        question (str): La question posée.

    Returns:
        str: La réponse trouvée dans le texte.
    """
    resultat = qa_pipeline(question=question, context=texte)
    return resultat['answer']

# 5. Exemple d'utilisation
if __name__ == "__main__":
    print("Réponses aux questions spécifiques à un texte financier avec FinBERT !")
    texte_financier = input("Entrez un texte financier : ")
    question = input("Posez une question basée sur le texte : ")

    if texte_financier.strip() == "" or question.strip() == "":
        print("Erreur : Texte ou question non fourni.")
    else:
        print("\nRéponse en cours...")
        reponse = repondre_a_question(texte_financier, question)

        # Afficher la réponse
        print("\nRéponse :")
        print(reponse)
```
